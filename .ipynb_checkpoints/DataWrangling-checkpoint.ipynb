```python
import pandas as pd
import missingno as msno
from sklearn import preprocessing
import re as re
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv("dataset/chemicals_disease.csv", encoding = "latin-1")
```

```python
#check dataframe is empty or not
dataframe.empty

```

```python
dataframe.info()
```

```python
msno.matrix(dataframe)
```

```python
dataframe.dropna(axis = "columns", inplace = True)
```

```python
msno.matrix(dataframe)
```

```python
dataframe
```

```python
dataframe["chemical_name"] = dataframe.chemical_name.apply(lambda x: re.sub('<[^<]+?>', '', x))
dataframe["disease_name"] = dataframe.disease_name.apply(lambda x: re.sub('<[^<]+?>', '', x))
dataframe
```

```python
dataframe.dtypes
```

```python
dataframe.groupby("disease_name").size()
```

```python
dataframe.groupby("verify_relationship").size()
```

```python
binarizer = preprocessing.LabelBinarizer()
dataframe["verify_relationship"] = binarizer.fit_transform(dataframe["verify_relationship"])
```

```python
dataframe["_unit_state"] = binarizer.fit_transform(dataframe["_unit_state"])
```

```python
dataframe.groupby("verify_relationship").size()
```

```python
dataframe.info()
```

Encode the categorical values

```python
columns = ["_unit_id","_unit_state","chemical_id","chemical_name","disease_id","disease_name","form_sentence","relation_pair_id","uniq_id"]
```

```python
le = preprocessing.LabelEncoder()

for x in columns:
    dataframe[x] = le.fit_transform(dataframe[x])


dataframe
```

```python
dataframe["_golden"] = dataframe["_golden"].astype(int)
```

```python
dataframe
```

```python
def draw_histograms(dataframe,columns, rows, cols):
    fig=plt.figure(figsize=(20,20))
    for i, feature in enumerate(columns):
        ax=fig.add_subplot(rows,cols,i+1)
        dataframe[feature].hist(bins=20,ax=ax,facecolor='midnightblue')
        ax.set_title(feature+" Distribution",color='Red')       
    fig.tight_layout()  
    plt.show()
draw_histograms(dataframe,dataframe.columns,6,3)
```

```python
dataframe.to_csv('dataset/cleaned_data.csv', sep=',')
```
